```python
train['Datetime'] = pd.to_datetime(train.Datetime, format = '%d-%m-%Y %H:%M')
test['Datetime'] = pd.to_datetime(test.Datetime, format = '%d-%m-%Y %H:%M')

train_original['Datetime'] = pd.to_datetime(train_original.Datetime, format = '%d-%m-%Y %H:%M')
test_original['Datetime'] = pd.to_datetime(test_original.Datetime, format = '%d-%m-%Y %H:%M')
```

```python
for df in (train, test, train_original, test_original):
    df['year'] = df.Datetime.dt.year
    df['month'] = df.Datetime.dt.month
    df['day'] = df.Datetime.dt.day
    df['hour'] = df.Datetime.dt.hour
```

```python
train['day_of_week'] = train['Datetime'].dt.dayofweek
temp = train['Datetime']
```

```python
def applyer(row):
    if row.dayofweek == 5 or row.dayofweek == 6:
        return 1
    else:
        return 0
    
temp2 = train['Datetime'].apply(applyer)
train['weekend'] = temp2
```

### Splitting data into training and validation part

```python
train_set = train.ix['2012-08-25' : '2014-06-24']
valid_set = train.ix['2014-06-25' : '2014-09-25']
```

```python
train_set.Count.plot(figsize = (15, 8), title = 'Daily Ridership', fontsize = 14, label = 'train')
valid_set.Count.plot(figsize = (15, 8), title = 'Daily Ridership', fontsize = 14, label = 'valid')
plt.xlabel('Datetime')
plt.ylabel('Passenger Count')
plt.legend(loc = 'best')
plt.show()
```

### Modelling Techniques

#### 1 - Naive Forecast

```python
dd = np.asarray(train_set.Count)
y_hat = valid_set.copy()
y_hat['naive'] = dd[len(dd) - 1]
plt.figure(figsize = (12, 8))
plt.plot(train_set.index, train_set.Count, label = "Train")
plt.plot(valid_set.index, valid_set.Count, label = "Valid")
plt.plot(y_hat.index, y_hat.naive, label = "Naive Forecast")
plt.legend(loc = 'best')
plt.title("Naive Forecast")
plt.show()
```

```python
# Calcualting RMSE to check the accuracy of our model
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(valid_set.Count, y_hat.naive))
print(rms)
```

#### From RMS it seen that this model is not suitable for datasets with high variability.

#### 2 - Moving Average

```python
y_hat_avg = valid_set.copy()
# Average of last 10 observations
y_hat_avg['moving_avg_forecast'] = train_set.Count.rolling(10).mean().iloc[-1]
plt.figure(figsize = (15, 5))
plt.plot(train_set.Count, label = 'train')
plt.plot(valid_set.Count, label='Valid') 
plt.plot(y_hat_avg.moving_avg_forecast, label='Moving Average Forecast using 10 observations') 
plt.legend(loc='best') 
plt.show() 

# Average of last 20 observations
y_hat_avg['moving_avg_forecast'] = train_set.Count.rolling(20).mean().iloc[-1]
plt.figure(figsize = (15, 5))
plt.plot(train_set.Count, label = 'train')
plt.plot(valid_set.Count, label='Valid') 
plt.plot(y_hat_avg.moving_avg_forecast, label='Moving Average Forecast using 20 observations') 
plt.legend(loc='best') 
plt.show() 

# Average of last 50 observations
y_hat_avg['moving_avg_forecast'] = train_set.Count.rolling(50).mean().iloc[-1]
plt.figure(figsize = (15, 5))
plt.plot(train_set.Count, label = 'train')
plt.plot(valid_set.Count, label='Valid') 
plt.plot(y_hat_avg.moving_avg_forecast, label='Moving Average Forecast using 50 observations') 
plt.legend(loc='best') 
plt.show() 
```

```python
# Calcualting RMSE to check the accuracy of Moving Average model
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(valid_set.Count, y_hat_avg.moving_avg_forecast))
print(rms)
```
